```python
from nltk.stem import WordNetLemmatizer 
lemmatizer = WordNetLemmatizer() 
```

```python
with open('data/iliada.txt', 'r') as file:
    text = file.read();
text
```

```python
text = text.lower()
pattern = r'[^a-zA-Z\s]'
text = re.sub(pattern, '', text)
text = [lemmatizer.lemmatize(x) for x in text.split()]
text = np.unique(text)
```

```python
np.random.shuffle(text)
```

```python
text = ' '.join(text)
text
```

```python
with open('data/iliada_processed.txt', 'w') as f:
    f.write(text)
```
